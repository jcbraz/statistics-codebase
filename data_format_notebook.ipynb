```python
untitled_columns_indexes = [i for i, col in enumerate(df.columns) if 'Untitled' in col]
untitled_columns_mapping = dict()

for i in untitled_columns_indexes:
    untitled_columns_mapping[df.columns[i-1]] = df.columns[i]

untitled_columns_mapping
```

```python
for key, value in untitled_columns_mapping.items():
    for i, row in df.iterrows():
        if row[key] == 'Other':
            df.at[i, key] = row[value]

df = df.drop(columns=untitled_columns_mapping.values())
df
```

```python
df['Which of the following options more accurately characterize your hobbies?'].value_counts()
```

```python
# Replace the nan values from the column 'What\'s your yearly income (40k = 40000€)' with 0
df['What\'s your yearly income (40k = 40000€)'] = df['What\'s your yearly income (40k = 40000€)'].fillna(0)
df
```

```python
amazon_prime_rows = df.loc[df['Do you have Amazon Prime subscription?'] == 'Yes']
amazon_prime_rows
```

```python
import numpy as np

amazon_prime_rows.loc[np.logical_and(amazon_prime_rows['Why do you think its worth it to have the Amazon Prime subscription? (1)'] == 'nan', amazon_prime_rows['Why do you think its worth it to have the Amazon Prime subscription? (1)'].isna())]
```

```python
for i, row in amazon_prime_rows.iterrows():
    if str(row['Why do you think its worth it to have the Amazon Prime subscription? (1)']).lower() == 'nan':
            if row['Why do you think its worth it to have the Amazon Prime subscription? (2)'] != None and row['Why do you think its worth it to have the Amazon Prime subscription? (2)'] != 'nan':
                if len(str(row['Why do you think its worth it to have the Amazon Prime subscription? (2)'])) > 1:
                    best_fit_value = str(row['Why do you think its worth it to have the Amazon Prime subscription? (2)']).split(',')[0]
                    df.at[i, 'Why do you think its worth it to have the Amazon Prime subscription? (1)'] = best_fit_value
                else:
                    best_fit_value = str(row['Why do you think its worth it to have the Amazon Prime subscription? (2)'])
                    df.at[i, 'Why do you think its worth it to have the Amazon Prime subscription? (1)'] = best_fit_value
    
df.rename(columns={'Why do you think its worth it to have the Amazon Prime subscription? (1)': 'Why do you think its worth it to have the Amazon Prime subscription?'}, inplace=True)
df.drop(columns=['Why do you think its worth it to have the Amazon Prime subscription? (2)'], inplace=True)
df
```
